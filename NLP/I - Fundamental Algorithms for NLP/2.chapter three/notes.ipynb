### O que é um modelo de trigramas?**

É um jeito de **tentar prever a próxima palavra** com base nas duas anteriores.

Exemplo:  
> Se eu já vi as palavras “eu gosto”, o modelo tenta **adivinhar** qual palavra vem depois.  
Talvez “de”, talvez “muito”, etc.

### Como calcula a chance da próxima palavra aparecer?**

Vê-se quantas vezes 3 palavras aparecem juntas (ex: "eu gosto de") e também quantas vezes só as duas primeiras aparecem ("eu gosto").

Então a fórmula é:

> **Chance da terceira palavra** =  
> Quantas vezes “eu gosto de” aparece  
> **dividido por**  
> Quantas vezes “eu gosto” aparece

### Exemplo prático**

Frase: `i want chinese food`  
(traduzindo: "eu quero comida chinesa")

Vai **multiplicando as chances** de cada palavra vir depois da outra:

- chance de “i” vir depois do começo da frase
- chance de “want” vir depois de “i”
- chance de “chinese” vir depois de “want”
- chance de “food” vir depois de “chinese”

Multiplica tudo e tem a **probabilidade** da frase.

### **Mas e se uma combinação nunca apareceu?**

Exemplo: nunca vi “chinese food” antes.  
Nesse caso a chance seria 0, ou seja, a frase “não pode acontecer”. Para resolver isso usa uma técnica chamada **Add-1 smoothing**:

> **Finge-se** que tudo apareceu **pelo menos uma vez**, mesmo que não tenha aparecido.

Assim, nenhuma frase vai ter chance 0.

### **Qual probabilidade vai ser maior?**

- Sem suavização → mais realista, mas arriscado (pode dar zero)
- Com suavização → mais segura, mas **menor**, porque distribui a "chance" com palavras que nunca apareceram.


### **Interpolação**

Junta-se:
- Um modelo que olha só **uma palavra antes**
- Com outro que olha só a **palavra sozinha**

Isso dá uma resposta **mais balanceada**.
